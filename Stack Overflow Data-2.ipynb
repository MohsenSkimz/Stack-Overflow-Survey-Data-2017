
## Q1       
### What are Top 5 Languages worked by professional?



```python
## Create a function that visualizes two columns of bar charts for top 5 languages worked by Student, Professional developer
def plot_top_languages(df, professional):
    '''
    INPUT:
    df - the pandas dataframe
    professional - visualize Student and Professional developer 

    OUTPUT:
    bar chart of top 5 languages worked by the professionals
    '''
    top_lang_professional = df["HaveWorkedLanguage"][df["Professional"] == professional].str.split(';', expand=True).stack().value_counts()
    fig_dims = (13,5)
    fig, ax = plt.subplots(figsize=fig_dims)
    sns.barplot(x=top_lang_professional[:5].index, y=top_lang_professional[:5].values).set(title='Top 5 Languages Worked by Professional  {}'.format(professional))

    plt.xlabel("Language", labelpad=15)
    plt.ylabel("Number of {}".format(professional), labelpad=15)

# Create a for loop that calls the function for each professional
for professional in ["Student","Professional developer"]:
    plot_top_languages(df, professional)
```

Like we see in frist bar chart for Student the top languages are JavaScript,python,java,SQL and PHP ,that Because theses languages are taught in University and schools , in other side for Professional developer the top languages are SQL,JavaScript,C#,python and PHP .

## Q2
### What are Most country that have People used JavaScript ?



```python
JavaScript_list = df[df["HaveWorkedLanguage"] == "JavaScript"]["Country"].str.split(';', expand=True)
JavaScript_list.head()
```

```python
JavaScript_country= JavaScript_list.stack().value_counts()
```

```python
JavaScript_country.head().plot(kind='bar', figsize=(10, 5), rot=0)
plt.xlabel("Country", labelpad=15)
plt.ylabel("Number of People", labelpad=15)
plt.title("Most Country That Have People Used JavaScript  ", y=1);
```

in the bar chart the most country that used JavaScript is United States then United kingdom then India 

## Q3
### What are Most IDE That Used By United States ?

```python
JavaScript_list = df[df["Country"] == "United States"]["IDE"].str.split(';', expand=True)
JavaScript_list.head()

```

```python
JavaScript_country= JavaScript_list.stack().value_counts()
```

```python
JavaScript_country.head().plot(kind='bar', figsize=(10, 5), rot=0)
plt.xlabel("IDE", labelpad=15)
plt.ylabel("Number of People", labelpad=15)
plt.title("Most IDE That Used By United States ", y=1);
```

So in this bar chart we see in this bar chart that most of the IDE that the US is using is Visual Studio .
